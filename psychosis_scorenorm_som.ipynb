```python
def build_dataset(dataset_list, dataset_id_to_name):

    X_data = np.concatenate(dataset_list, axis=0)
    y_labels = np.concatenate([np.full(len(d), i) for i, d in enumerate(dataset_list)])
    assert len(y_labels) == len(X_data)

    mu, std = np.mean(X_data[y_labels == 0], axis=0), np.std(X_data[y_labels == 0], axis=0)
    X_data = (X_data - mu) / std

    train_data = X_data[y_labels < 2]
    train_labels =  y_labels[y_labels < 2]


    test_data = X_data[y_labels > 1]
    test_labels = y_labels[y_labels > 1]
    print(y_labels)
    return (train_data, train_labels), (test_data, test_labels)
```

```python
def build_bmu_df(net, test_data, test_labels):
    
    bmus_arr = net.find_bmu_ix(np.asarray(test_data))
    bmu_dists = [np.linalg.norm(net.nodes_list[bmuidx].weights - x) for bmuidx,x in zip(bmus_arr, test_data)]

    bmus = pd.DataFrame([b.item() for b in bmus_arr], columns=["BMU"])
    bmus['dataset'] = [dataset_id_to_name[label] for label in test_labels]
    bmus['bmu-distance'] = bmu_dists
    bmu_df = bmus.set_index('BMU')

    for idx in bmu_df.index:
        node = net.nodes_list[idx]
        bmu_df.loc[idx, ['x', 'y']] = node.pos
    
    bmucounts = bmus[['BMU', 'dataset']].groupby(['BMU']).value_counts().reset_index(level=-1, name='count')
    bmudists = bmus[['BMU', 'bmu-distance']].groupby(['BMU']).mean()
    bmudists_median = bmus[['BMU', 'bmu-distance']].groupby(['BMU']).median()
    
    for idx in bmucounts.index:
        node = net.nodes_list[idx]
        bmucounts.loc[idx, ['x', 'y']] = node.pos
        bmucounts.loc[idx, 'neighbourhood-distance'] = node.difference
        bmucounts.loc[idx, 'mean-matching-distance'] = bmudists.loc[idx, 'bmu-distance']        

    significant_bmus = []
    dataset_counts = bmus['dataset'].value_counts()
    
    for name, g in bmucounts.groupby('dataset'):
        cohort_count = dataset_counts[name]
        #min_cohort_count = int(0.0*cohort_count)
        significant_bmus.append(g[g['count'] >= int(0.00*cohort_count)])

    significant_bmus = pd.concat(significant_bmus).sort_values('count', ascending=False)
    
    return bmu_df, significant_bmus
```

```python
def build_hexbin_source(net, train_data, bmu_df, significant_bmus):
    _x,_y = bmu_df.x, bmu_df.y
    _q,_r = cartesian_to_axial(_x,_y, size=HEXSIZE, orientation="pointytop", aspect_scale=1)
    _sx,_sy = axial_to_cartesian(_q,_r,  size=1, orientation="pointytop")

    scattersource = ColumnDataSource(dict(x= _sx, y=_sy, dataset=bmu_df['dataset'],
                                    color=[color_dict[d] for d in bmu_df['dataset']]
    ))
    
    bmu_density = net.find_bmu_ix(np.asarray(train_data))
    bmu_density = Counter(bmu_density)
    
    x,y,distance, sample_count, density = [],[],[],[], []
    binned_density = []
    for idx, n in enumerate(net.nodes_list):
        x.append(n.pos[0].get())
        y.append(n.pos[1].get())
        distance.append(n.difference)
        c = bmu_density.get(idx, 1e-5)
        d = 100 * c / len(train_data)
        density.append(d)
        binned_density.append(np.ceil(d) if d > 1e-5 else 0)
        
        # add an average of matching distances? train distances or test distances?
        
    # This defines the datapoints for the heatmap background
    x,y,distance = np.asarray(x),np.asarray(y),np.asarray(distance)
    q,r = cartesian_to_axial(x,y, size=HEXSIZE, orientation="pointytop")
    binsource = ColumnDataSource(dict(q=q,r=r, distance=distance,
                                      density=density, binned_density=binned_density))
    
    # This defines the source of the scatter plots of test samples
    # to be overlayed on the heatmap
    _x,_y = bmu_df.x, bmu_df.y
    _q,_r = cartesian_to_axial(_x,_y, size=HEXSIZE, orientation="pointytop", aspect_scale=1)
    sx, sy = axial_to_cartesian(_q,_r,  size=1, orientation="pointytop")
    scattersource = ColumnDataSource(dict(x= _sx, y=_sy, dataset=bmu_df['dataset'],
                                    color=[color_dict[d] for d in bmu_df['dataset']]
    ))
    
    # Optionally we can plot only the significant nodes (i.e nodes w/ >= 10% of the dataset)
    x = significant_bmus.x
    y = significant_bmus.y
    dataset = significant_bmus['dataset']
    count = significant_bmus['count']
    q,r = cartesian_to_axial(x,y, size=HEXSIZE, orientation="pointytop", aspect_scale=1)
    sx,sy = axial_to_cartesian(q,r,  size=1, orientation="pointytop")
    significantsource = ColumnDataSource(dict(x=sx,y=sy, dataset=dataset,
                                              color=[color_dict[d] for d in dataset],
                                              scale= scale(significant_bmus['count'], 10, 50)))
    
    return binsource, scattersource, significantsource
```

```python
# Some plotting defaults

HEXSIZE = .99 / np.sqrt(3)
marker_dict = OrderedDict({"Down Syndrome":"triangle", "ASD":"star", "HR-Typical":"circle", "Atypical":"diamond"})
DATASETS = list(marker_dict.keys())
MARKERS = list(marker_dict.values())
color_dict = OrderedDict({"Down Syndrome":'#f03b20', #red
                          "HR-Typical":"#7fc97f", #green
                          "ASD":"#fdc076", #orange
                          "Atypical":"#fdc086", #blue
                         })

SCALE_FACTOR = 1
background_fill_color="#fafafa"
header_text_size = "18px"
MARKER_SIZE = 13

def plot_hex_heatmap(scattersource, binsource, heatmap_col='distance',
                     cmap_upper_ratio = 1.02, cmap_lower_ratio=1, 
                     width=800, height=700, cmap='Greys256', cbar_ticks=True, scatter_jitter=True):

    
#     group_filter = BooleanFilter([c != "LR-Typical" for c in bmusdf['dataset']])
#     view = CDSView(filter=group_filter)
    heatmap_min, heatmap_max = min(binsource.data[heatmap_col]), max(binsource.data[heatmap_col])
    
    
    plot = figure(width=int(width * SCALE_FACTOR), height=int(height * SCALE_FACTOR), background_fill_color=background_fill_color)
    r = plot.hex_tile(
        source=binsource, color="darkgray",
        fill_color=linear_cmap(heatmap_col, cmap,
                               low=cmap_lower_ratio*heatmap_min,
                               high=cmap_upper_ratio*heatmap_max),
        alpha=0.8
    )
    
    if cbar_ticks:
        cbar_font_size=f"{12*SCALE_FACTOR}pt"
        cbar_tick_color='black'
    else:
        cbar_font_size="0pt"
        cbar_tick_color=None
    
    tick_values = np.arange(1,6)
    tick_offset = 0.04  # Adjust based on how much offset you need
    offset_tick_values = [tick + tick_offset for tick in tick_values]
    tick_labels = {k:str(int(k)) for k in offset_tick_values}
    
    
    color_bar = r.construct_color_bar(padding=1, title="Percentage of Typicals",
                                      title_text_font_size=f"{18*SCALE_FACTOR}px",
                                     major_label_text_font_size=cbar_font_size,
                                     major_tick_line_color=cbar_tick_color,
                                    ticker=FixedTicker(ticks=offset_tick_values),
                                      major_tick_line_width=2*SCALE_FACTOR,
                                     )
    color_bar.major_label_overrides = tick_labels
    plot.add_layout(color_bar, "right")

    plot.scatter(source=scattersource,
                 marker=factor_mark('dataset', MARKERS, DATASETS),
                 color='color',
                 x=jitter('x', .8) if scatter_jitter else 'x',
                 y=jitter('y', .8) if scatter_jitter else 'y',
                 size='scale' if 'scale' in scattersource.data else MARKER_SIZE*SCALE_FACTOR,
                 legend_group="dataset",
                 alpha=0.9,
                 line_color='black'
    )

    legend = Legend(
        items=plot.legend.items,
        location='top_right',
        orientation='horizontal',
        title='Dataset',
        title_text_font_size=f"{int(14*SCALE_FACTOR)}px",
        background_fill_color=background_fill_color,
        label_text_font_size=f"{int(13*SCALE_FACTOR)}px",
        glyph_height=int(32*SCALE_FACTOR),
        glyph_width=int(32*SCALE_FACTOR)
    )

    # Remove original legend
    plot.legend.destroy()
    plot.add_layout(legend, 'above')

    plot.axis.visible = False
    plot.grid.visible = False
    
    return plot
```

```python
import os
import numpy as np
import pandas as pd
from scipy.stats import spearmanr
from statsmodels.stats.multitest import multipletests
import warnings

# —————————————————————————————————
#  CONFIG
# —————————————————————————————————
CORRECTION   = "bonf"              # "bonf" or "fdr"
alpha_values = {"bonf": 0.05, "fdr": 0.05}
SCORE_SOURCE = "gmm"               # "score_norms", "gmm", or "kd"

outdir = os.path.join(workdir, f"allatypicals_spearman_{CORRECTION}_{SCORE_SOURCE}")
os.makedirs(outdir, exist_ok=True)

# —————————————————————————————————
#  BUILD SCORE DF
# —————————————————————————————————
if SCORE_SOURCE == "score_norms":
    # 1) pull in all of your autoencoder score‐dims
    score_cols = [f"score_dim_{i}" for i in range(outliers.shape[1])]
    score_df   = pd.DataFrame(outliers, columns=score_cols)
    # 2) stitch on the subject IDs so we know who is who
    score_df["src_subject_id"] = outlier_subject_ids
    # 3) now merge in the BMU for each of those same subjects
    score_df   = score_df.merge(
        pdp_df[["src_subject_id","bmu"]],
        on="src_subject_id", how="inner"
    )

elif SCORE_SOURCE in ("gmm","kd"):
    # read the CSV you already have
    tmp       = pd.read_csv(os.path.join(workdir, f"mixed_all_features_{SCORE_SOURCE}_scores.csv"))
    tmp       = tmp.rename(columns={"subject_id":"src_subject_id"})
    score_col = f"{SCORE_SOURCE}_ood_score"
    # pick just src_subject_id + that one ood‐score
    score_df  = tmp[["src_subject_id", score_col]]
    # then merge in the BMU
    score_df  = score_df.merge(
        pdp_df[["src_subject_id","bmu"]],
        on="src_subject_id", how="inner"
    )
    score_cols = [score_col]

else:
    raise ValueError(f"Unknown SCORE_SOURCE: {SCORE_SOURCE}")

# quick sanity check: we should now have exactly your subjectIDs + bmu + score_cols
print("→ score_df cols:", score_df.columns.tolist())

# —————————————————————————————————
#  LOAD BEHAVIOR
# —————————————————————————————————
behavior_df   = pd.read_csv(os.path.join(workdir, "behavior_cognition_not.csv"))
behavior_cols = [
    'pps_y_ss_number', 
    'pps_y_ss_bother_sum', 
    #'cbcl_scr_syn_anxdep_r', 
    #'cbcl_scr_syn_withdep_r', 
    #'cbcl_scr_syn_somatic_r', 
    #'cbcl_scr_syn_social_r', 
    #'cbcl_scr_syn_thought_r', 
    #'cbcl_scr_syn_attention_r', 
    #'cbcl_scr_syn_rulebreak_r',
    #'cbcl_scr_syn_aggressive_r', 
    #'cbcl_scr_syn_internal_r', 
    #'cbcl_scr_syn_external_r', 
    #'cbcl_scr_syn_totprob_r', 
    #'cbcl_scr_dsm5_depress_r', 
    #'cbcl_scr_dsm5_anxdisord_r', 
    #'cbcl_scr_dsm5_somaticpr_r', 
    #'cbcl_scr_dsm5_adhd_r', 
    #'cbcl_scr_dsm5_opposit_r', 
    #'cbcl_scr_dsm5_conduct_r', 
    #'cbcl_scr_07_sct_r', 
    #'cbcl_scr_07_ocd_r', 
    #'cbcl_scr_07_stress_r'
]
# —————————————————————————————————
#  MERGE
# —————————————————————————————————
merged = score_df.merge(behavior_df, on="src_subject_id", how="inner")
# — make sure `merged` has a 'bmu' column —  
# pdp_df is the one you built earlier with ['src_subject_id','bmu','label']


# sanity‐check:
print("Merged columns:", merged.columns.tolist())

n_subj = len(merged)
if n_subj==0:
    raise RuntimeError("No overlap between scores ↔ behavior!")

# —————————————————————————————————
#  SPEARMAN LOOP
# —————————————————————————————————
recs = []
with warnings.catch_warnings():
    warnings.simplefilter("ignore", category=UserWarning)

    if SCORE_SOURCE=="score_norms":
        for dim in score_cols:
            x=merged[dim]
            if x.nunique()<=1: continue
            for beh in behavior_cols:
                y=merged[beh]
                if y.nunique()<=1: continue
                rho,p= spearmanr(x,y, nan_policy="omit")
                recs.append({
                  "score_dim": dim, "behavior":beh,
                  "rho":rho, "pval":p,
                  "n_subjects":n_subj,
                  "score_mean":x.mean(),"score_std":x.std(),
                  "behav_mean":y.mean(),"behav_std":y.std()
                })
    else:
        x=merged[score_cols[0]]
        if x.nunique()>1:
            for beh in behavior_cols:
                y=merged[beh]
                if y.nunique()<=1: continue
                rho,p= spearmanr(x,y, nan_policy="omit")
                recs.append({
                  "score_dim": score_cols[0], "behavior":beh,
                  "rho":rho, "pval":p,
                  "n_subjects":n_subj,
                  "score_mean":x.mean(),"score_std":x.std(),
                  "behav_mean":y.mean(),"behav_std":y.std()
                })

# —————————————————————————————————
#  CORRECT + DUMP CSVs
# —————————————————————————————————
if not recs:
    print("No correlations computed.")
else:
    dfc = pd.DataFrame(recs)
    method = {"bonf":"bonferroni","fdr":"fdr_bh"}[CORRECTION]
    alpha  = alpha_values[CORRECTION]

    rej, p_corr, _, _ = multipletests(dfc["pval"], alpha=alpha, method=method)
    dfc[f"p_{CORRECTION}"]=p_corr
    dfc["significant"]=rej

    # round floats
    for c in dfc.select_dtypes("float"):
        dfc[c]=dfc[c].round(3)

    # full
    dfc.to_csv(os.path.join(outdir,f"aggregate_{SCORE_SOURCE}_full.csv"),index=False)
    print("✔ saved full")

    # significant only
    sig = dfc[dfc["significant"]].copy()
    if len(sig):
        sig["abs_rho"]=sig["rho"].abs()
        sig = sig.sort_values("abs_rho",ascending=False).drop(columns="abs_rho")
        sig.to_csv(os.path.join(outdir,f"aggregate_{SCORE_SOURCE}_{CORRECTION}_significant.csv"),index=False)
        print(f"✔ saved {CORRECTION} significant")
    else:
        print(f"No {CORRECTION}‑significant results")

    # raw p<0.05
    unc = dfc[dfc["pval"]<0.05]
    if len(unc):
        unc.to_csv(os.path.join(outdir,f"aggregate_{SCORE_SOURCE}_uncorrected_p05.csv"),index=False)
        print("✔ saved raw p<0.05")
    else:
        print("No raw p<0.05")

print("✅ Done.")
```

```python
# =============================================================================
# SPEARMAN CORRELATION BY BMU (SAVE ALL + SIGNIFICANT RESULTS, FDR vs BONF)
#   — now rounds all float columns to 3 decimals before saving
# =============================================================================

import os
import numpy as np
import pandas as pd
from scipy.stats import spearmanr
from statsmodels.stats.multitest import multipletests
from collections import defaultdict
import warnings

# -----------------------------------------------------------------------------
# 1) CONFIGURE & PARAMS
# -----------------------------------------------------------------------------
min_subjects_per_bmu = 25

# choose multiple-testing correction: "fdr" or "bonf"
CORRECTION = "bonf"
alpha      = 0.05

SCORE_SOURCE     = "gmm"  # "score_norms", "gmm", or "kd"
INCLUDE_INLIERS  = False

outdir = os.path.join(workdir, f"spearman_{CORRECTION}_{SCORE_SOURCE}_{som_height}x{som_width}")
os.makedirs(outdir, exist_ok=True)
sig_dir = os.path.join(outdir, "sig")
os.makedirs(sig_dir, exist_ok=True)

print(f"[INFO] SCORE_SOURCE = {SCORE_SOURCE}")
print(f"[INFO] CORRECTION = {CORRECTION.upper()} @ α={alpha}")
print(f"[INFO] Output directory = {outdir}")
print(f"[INFO] SOM size: {som_height}x{som_width}")

# -----------------------------------------------------------------------------
# 2) BUILD score_df
# -----------------------------------------------------------------------------
if SCORE_SOURCE == "score_norms":
    cols = [f"score_dim_{i}" for i in range(outliers.shape[1])]
    score_df = pd.DataFrame(outliers, columns=cols)
    score_df["src_subject_id"] = outlier_subject_ids
    if INCLUDE_INLIERS:
        in_df = pd.DataFrame(abcd_inlier_test_data, columns=cols)
        in_df["src_subject_id"] = abcd_inlier_test_subject_ids
        score_df = pd.concat([score_df, in_df], ignore_index=True)

elif SCORE_SOURCE in ("gmm", "kd"):
    csv_name = f"mixed_all_features_{SCORE_SOURCE}_scores.csv"
    tmp = pd.read_csv(os.path.join(workdir, csv_name)) \
             .rename(columns={"subject_id": "src_subject_id"})
    score_col = f"{SCORE_SOURCE}_ood_score"
    score_df = tmp[["src_subject_id", score_col]]
    cols = [score_col]

else:
    raise ValueError(f"Unknown SCORE_SOURCE: {SCORE_SOURCE}")

print(f"[INFO] Score data → {len(score_df)} rows.")

# -----------------------------------------------------------------------------
# 3) MERGE WITH BEHAVIORAL DATA
# -----------------------------------------------------------------------------
behavior_df = pd.read_csv(os.path.join(workdir, "behavior_cognition_not.csv"))
behavior_cols = [
    'pps_y_ss_number', 
    'pps_y_ss_bother_sum', 
    #'cbcl_scr_syn_anxdep_r', 
    #'cbcl_scr_syn_withdep_r', 
    #'cbcl_scr_syn_somatic_r', 
    #'cbcl_scr_syn_social_r', 
    #'cbcl_scr_syn_thought_r', 
    #'cbcl_scr_syn_attention_r', 
    #'cbcl_scr_syn_rulebreak_r',
    #'cbcl_scr_syn_aggressive_r', 
    #'cbcl_scr_syn_internal_r', 
    #'cbcl_scr_syn_external_r', 
    #'cbcl_scr_syn_totprob_r', 
    #'cbcl_scr_dsm5_depress_r', 
    #'cbcl_scr_dsm5_anxdisord_r', 
    #'cbcl_scr_dsm5_somaticpr_r', 
    #'cbcl_scr_dsm5_adhd_r', 
    #'cbcl_scr_dsm5_opposit_r', 
    #'cbcl_scr_dsm5_conduct_r', 
    #'cbcl_scr_07_sct_r', 
    #'cbcl_scr_07_ocd_r', 
    #'cbcl_scr_07_stress_r'
]
merged = pd.merge(score_df, behavior_df, on="src_subject_id", how="inner")
print(f"[INFO] Merged → {len(merged)} subjects, behaviors: {behavior_cols}")

# -----------------------------------------------------------------------------
# 4) LOAD BMU → IDs
# -----------------------------------------------------------------------------
bmu_to_ids = defaultdict(list)
with open(os.path.join(workdir, "pdp_bmu_id_sets.txt")) as f:
    for line in f:
        line = line.strip()
        if not line: continue
        bmu = int(line.split(":",1)[0].replace("BMU","").strip())
        ids = [x.strip() for x in line.split(":",1)[1].split(",") if x.strip()]
        bmu_to_ids[bmu].extend(ids)
print(f"[INFO] Loaded {len(bmu_to_ids)} BMU mappings.")

# -----------------------------------------------------------------------------
# 5) PER-BMU SPEARMAN + COLLECT ALL RESULTS
# -----------------------------------------------------------------------------
all_recs = []
all_sig  = []
method   = "fdr_bh" if CORRECTION=="fdr" else "bonferroni"

with warnings.catch_warnings():
    warnings.simplefilter("ignore", category=UserWarning)

    for bmu, ids in sorted(bmu_to_ids.items()):
        sub = merged[merged["src_subject_id"].isin(ids)]
        n   = len(sub)
        if n < min_subjects_per_bmu:
            print(f"[SKIP] BMU {bmu}: only {n} subjects")
            continue

        print(f"[PROC] BMU {bmu}: {n} subjects")
        recs = []

        if SCORE_SOURCE == "score_norms":
            for dim in cols:
                x = sub[dim]
                if x.nunique() <= 1: continue
                for beh in behavior_cols:
                    y = sub[beh]
                    if y.nunique() <= 1: continue
                    rho, p = spearmanr(x, y, nan_policy="omit")
                    recs.append({
                        "bmu": bmu,
                        "score_dim": dim,
                        "behavior": beh,
                        "rho": rho,
                        "pval": p,
                        "n": n,
                    })
        else:
            dim = cols[0]
            x   = sub[dim]
            if x.nunique() > 1:
                for beh in behavior_cols:
                    y = sub[beh]
                    if y.nunique() <= 1: continue
                    rho, p = spearmanr(x, y, nan_policy="omit")
                    recs.append({
                        "bmu": bmu,
                        "score_dim": dim,
                        "behavior": beh,
                        "rho": rho,
                        "pval": p,
                        "n": n,
                    })

        if not recs:
            print(f"[INFO] BMU {bmu}: no valid correlations")
            continue

        df_bmu = pd.DataFrame(recs)
        # round to 3 decimals
        float_cols       = df_bmu.select_dtypes(include="float").columns
        df_bmu[float_cols] = df_bmu[float_cols].round(3)

        all_recs.append(df_bmu)

        # apply correction
        reject, p_corr, _, _ = multipletests(df_bmu["pval"], alpha=alpha, method=method)
        df_bmu[f"p_{CORRECTION}"] = p_corr.round(3)
        df_bmu["significant"]    = reject

        sig = df_bmu[df_bmu["significant"]].copy()
        if not sig.empty:
            all_sig.append(sig)
        print(f"[INFO] BMU {bmu}: {len(sig)} significant (by {CORRECTION})")

# -----------------------------------------------------------------------------
# 6) AGGREGATE & SAVE
# -----------------------------------------------------------------------------
if all_recs:
    agg_full = pd.concat(all_recs, ignore_index=True)
    # round again to catch any new floats
    float_cols      = agg_full.select_dtypes(include="float").columns
    agg_full[float_cols] = agg_full[float_cols].round(3)
    full_path = os.path.join(outdir, f"aggregate_all_{CORRECTION}.csv")
    agg_full.to_csv(full_path, index=False)
    print(f"[SAVED] All correlations → {full_path}")

if all_sig:
    agg_sig = pd.concat(all_sig, ignore_index=True)
    float_cols      = agg_sig.select_dtypes(include="float").columns
    agg_sig[float_cols] = agg_sig[float_cols].round(3)
    sig_path = os.path.join(sig_dir, f"aggregate_sig_{CORRECTION}.csv")
    agg_sig.to_csv(sig_path, index=False)
    print(f"[SAVED] Significant correlations → {sig_path}")

# raw uncorrected
if all_recs:
    unc = agg_full[agg_full["pval"] < 0.05]
    if not unc.empty:
        unc_path = os.path.join(outdir, "aggregate_uncorrected_p05.csv")
        unc.to_csv(unc_path, index=False)
        print(f"[SAVED] Uncorrected p<0.05 → {unc_path}")
    else:
        print("[RESULT] No uncorrected p<0.05 correlations.")
```

```python
# =============================================================================
# Per‑BMU Pearson scatterplots + annotate y=mx+b + pick top‑N by |slope|
# 
# =============================================================================
import os
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import shutil
from scipy.stats import pearsonr

# ──────────────────────────────────────────────────────────────────────────────
# prerequisites (from your BMU‑correlation cell):
#   merged, cols (score columns), behavior_cols, bmu_to_ids,
#   outdir, min_subjects_per_bmu, SCORE_SOURCE, som_height, som_width
# ──────────────────────────────────────────────────────────────────────────────

# how many “top” slopes to keep per BMU
top_n = 20 if SCORE_SOURCE == "score_norms" else 5

# base folder for per‑BMU scatterplots
scatter_base = os.path.join(
    outdir,
    f"per_bmu_scatterplots_pearson_{som_height}x{som_width}"
)
os.makedirs(scatter_base, exist_ok=True)
sns.set_style("whitegrid")

for bmu, ids in sorted(bmu_to_ids.items()):
    sub = merged[merged["src_subject_id"].isin(ids)]
    n   = len(sub)
    if n < min_subjects_per_bmu:
        continue  # skip small BMUs

    # create BMU folder + most_significant subfolder
    bmu_dir = os.path.join(scatter_base, f"BMU_{bmu}")
    sig_dir = os.path.join(bmu_dir, "most_significant")
    os.makedirs(bmu_dir, exist_ok=True)
    os.makedirs(sig_dir, exist_ok=True)

    slopes_info = []  # will hold (|slope|, filepath)

    for score in cols:
        if score not in sub.columns:
            continue
        # subfolder per score_dim if present
        score_dir = os.path.join(bmu_dir, score) if SCORE_SOURCE=="score_norms" else bmu_dir
        os.makedirs(score_dir, exist_ok=True)

        for beh in behavior_cols:
            if beh not in sub.columns:
                continue
            data = sub[[beh, score]].dropna()
            if len(data) < 5 or data[beh].nunique() < 2:
                continue

            x = data[beh].values
            y = data[score].values
            # compute slope/intercept
            m, b = np.polyfit(x, y, 1)
            # Pearson r and p‑value
            r, pval = pearsonr(x, y)

            # draw scatter + fit
            fig, ax = plt.subplots(figsize=(8, 6))
            sns.regplot(
                data=data, x=beh, y=score, ax=ax,
                scatter_kws={"s": 40, "alpha": 0.6, "edgecolor": "w"},
                line_kws={"color": "crimson", "lw": 2},
            )
            # annotate with 3‑decimal precision
            ax.text(
                0.05, 0.95,
                f"y = {m:.3f}x + {b:.3f}\nr = {r:.3f}, p = {pval:.3f}",
                transform=ax.transAxes, fontsize=12, va="top"
            )
            ax.set_xlabel(beh, fontsize=12)
            ax.set_ylabel(score, fontsize=12)
            ax.set_title(f"BMU {bmu}: {score} vs. {beh}", fontsize=14)
            plt.tight_layout()

            # save figure
            fname    = f"{score}_vs_{beh}.png"
            out_path = os.path.join(score_dir, fname)
            fig.savefig(out_path, dpi=150, bbox_inches="tight")
            plt.close(fig)

            slopes_info.append((abs(m), out_path))

    # copy only the top‑N by |slope| into most_significant/
    for rank, (_, src) in enumerate(sorted(slopes_info, key=lambda x: x[0], reverse=True)[:top_n], start=1):
        dst = os.path.join(sig_dir, os.path.basename(src))
        shutil.copy(src, dst)
        print(f"[BMU {bmu} TOP {rank}] {os.path.basename(src)} | slope={os.path.basename(src).split('_vs_')[0]}={slopes_info[rank-1][0]:.3f}")
print("Done")
```
